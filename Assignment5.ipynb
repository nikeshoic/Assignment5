```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
car_data = pd.read_csv('../CarPrice.csv')
```

```python
car_data.info()
```

```python
car_data.select_dtypes(include=['int64','float64']).columns
```

```python
list = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'boreratio','stroke', 
        'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']
```

```python
lr = LinearRegression()
```

```python
y_pred_list = []
```

```python
y = car_data[['price']]
for value in list:
    x = car_data[[value]]

    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=42)

    lr.fit(x_train,y_train)
    print(value.upper())
    print(f"Coef : {lr.coef_}")
    print(f"Intercept: {lr.intercept_}")
    

    y_pred = lr.predict(x_test)
    y_pred_list.append(y_pred)
    
    mse = mean_squared_error(y_pred, y_test)
    r2score = r2_score(y_pred, y_test.values)
    
    print(f"Mean Squared Error: {mse}")
    print(f"R2 Score: {r2score}")

    plt.scatter(x_test,y_test)
    plt.plot(x_test,y_pred,c='r')
    plt.show()
    print()
    
```

```python
i=0
for value in y_pred_list:
    print(list[i].upper())
    print(value)
    print()
    i += 1
```
